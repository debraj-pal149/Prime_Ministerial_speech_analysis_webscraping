```python
from bs4 import BeautifulSoup
import requests
import datetime
speechfile = open("Speeches.html", 'r', encoding='utf-8')
code = speechfile.read()
print(code)
```

```python
main=BeautifulSoup(code,'html.parser')
main.prettify()
```

```python
ttt=main.find_all('span',"date")
print("Total Speeches : " + str(len(ttt)))
ttt
```

```python
urls=[]
for hh in main.find_all('h3'):
    if hh.find('a')!=None:
        a=hh.find('a')
        urls.append(a.attrs['href'])
urls
```

```python
dictt={}
for i in range(len(urls)):
    dt_str=ttt[i].text
    dt_obj=datetime.datetime.strptime(dt_str,'%b %d, %Y')
    dictt[dt_obj.date()]=urls[i]
dictt
```

```python
countdict={}
for q in range(len(dictt)):
    URL=list(dictt.values())[q]
    r=requests.get(URL)
#     print(r.content)
    soup=BeautifulSoup(r.content,'html.parser')
   
    t=soup.find_all('p')
    s=''
    for i in range(len(t)):
        if t[i].attrs=={'lang': 'hi', 'dir': 'ltr'}:
            break
        s=s+t[i].text
    # print(s)
    s=s.upper()
    wordlist=s.split()
    # print(wordlist)
    c=wordlist.count('ECONOMY')+wordlist.count('ECONOMIC')
    countdict[list(dictt.keys())[q]]=c
countdict
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
plt.figure(figsize=(16,5))
dates=matplotlib.dates.date2num(list(countdict.keys()))
matplotlib.pyplot.plot_date(dates,countdict.values())
# dates
```
